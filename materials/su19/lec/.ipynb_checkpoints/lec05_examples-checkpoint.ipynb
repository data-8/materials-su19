```python
from datascience import *
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
```

## Lecture 5  Part 2 - Examples

```python
heroes = make_array('Batman', 'Aquaman', 'Spiderman', 'Captain Marvel','Wonder Woman')
heroes
```

```python
superheroes = Table().with_columns(
    'Name', heroes,
    'Coolness', np.arange(5)
)

superheroes
```

```python
superheroes.where("Coolness", are.equal_to(4))
```

```python
superheroes.where("Coolness", 4)
```

```python
marvel = make_array(True, False, True, True, False)
superheroes = superheroes.with_column("Marvel Property", marvel)
superheroes
```

```python
superheroes.where("Coolness", are.above(0)).num_rows
```

```python
np.count_nonzero(superheroes.column("Coolness"))
```

```python
# True is considered 1, False is considered 0
np.count_nonzero(superheroes.column("Marvel Property"))
```

## Census ##

```python
#onscreeen
full = Table.read_table('nc-est2014-agesex-res.csv')
full
```

```python
partial = full.select('SEX', 'AGE', 'POPESTIMATE2010', 'POPESTIMATE2014')
partial
```

```python
simple = partial.relabeled('POPESTIMATE2010', '2010').relabeled('POPESTIMATE2014','2014')
simple
```

```python
simple.sort('AGE', descending=True)
```

## Visualization ##

```python
no_999 = simple.where('AGE', are.below(999))
```

```python
everyone = no_999.where('SEX', 0).drop('SEX')
```

```python
everyone.plot('AGE', '2010')
```
